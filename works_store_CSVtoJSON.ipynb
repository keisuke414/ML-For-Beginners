```python
from google.colab import drive
drive.mount('/content/drive')
from bs4 import BeautifulSoup                                                                                            
import requests, json
import pandas as pd
import glob
import codecs
import numpy as np
import re
import csv
import time
import os
import pathlib
import datetime
import collections
import shutil
import itertools
import urllib.request
import collections
```

```python
data_path = '/content/drive/MyDrive/store/csv/'

result_L_L = sorted(glob.glob(data_path + '*total_result.csv'))
result_L_L = sorted(result_L_L, key=lambda x:int((re.search(r"[0-9]+", x)).group(0)))
for result_L_L_i in result_L_L:
  this_pref_num = re.sub(r'total_result.csv', '', result_L_L_i.split('/')[-1])
  print(this_pref_num)
  print(result_L_L_i)
  # CSV ファイルの読み込み
  with open(result_L_L_i, 'r') as f:
    for row in csv.DictReader(f):
      json_list = []
      json_list.append(row)
      print(row['id'])
      # JSON ファイルへの書き込み
      save_path_dir = data_path + 'json/' + this_pref_num + '/'
      if not os.path.exists(save_path_dir):
        os.mkdir(save_path_dir)
      if not '/' in row['id']:
        with open(save_path_dir + row['id'] + '.json', 'w') as f:
          json.dump(json_list[0], f, ensure_ascii=False)
```
